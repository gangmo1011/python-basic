```
adut_female column을 만들고 나이(age)가 20이상이고 성별(sex)이 'female'인 값은 True를, 아니면 False 값을 할당합니다.
```

````python
import seaborn as sns
import pandas as pd
import numpy as np

titanic=sns.load_dataset('titanic')
titanic.count() # 카운트는 nan값을 제외하고 갯수를 셈
````

````python
# dropna로 풀기 
titanic.dropna(subset=['age'],axis=0,inplace=True)

titanic['adult_female']=titanic.apply(lambda x: True if (x.age >=20)&(x.sex=='female') else False, axis=1)

titanic
````

````python
# 인덱싱으로 풀기
titanic[titanic.age.isna()].index
````

````python
titanic.drop(index=titanic[titanic.age.isna()].index, axis=0,inplace=True)

titanic['adult_female']=titanic.apply(lambda x: True if (x.age >=20)&(x.sex=='female') else False, axis=1)

titanic
````

# bar 차트
- bar 명령과 barh 명령

````python
y=[2,3,1]
x=['가','나','다']
plt.bar(x,y)
plt.xlabel('가나다')
plt.ylabel('빈도수')
plt.show()
````

- color 키워드 인수로 색상 리스트 전달 가능

````python
y=[2,3,1]
x=['가','나','다']
colors=['red','green','blue']
plt.bar(x,y,color=colors)
plt.xlabel('가나다')
plt.ylabel('빈도수')
plt.show()
````

- xerr 인수나 yerr 인수를 지정하면 에러바(error bar)를 추가 가능
- alpha는 투명도를 지정. 0이면 완전 투명, 1이면 완전 불투명

````python
np.random.seed(0)

people=['몽','춘','방','향']
performance= 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title('Barh Chart')
plt.barh(people, performance, xerr=error, alpha = 0.4)
plt.xlabel('x 라벨')
plt.show()
````

### 연습 문제

````python
x=['사망자','생존자']
y=[len(titanic[titanic.alive == 'no']),len(titanic[titanic.alive == 'yes'])]

plt.barh(x,y,color=['red','green'],alpha=0.7)
plt.title('타이타닉호 생존 여부')
plt.xlabel('인원 수')
````

# Pie 차트 

````python
labels=['월세','전세','자가','기타']
sizes=[15,30,45,10]
colors=['yellowgreen','gold','lightskyblue','lightcoral']
explode=(0, 0, 0.1, 0)
plt.title('Pie Chart')
plt.pie(sizes,explode=explode, labels=labels, colors=colors,
        autopct='%1.lf%%', shadow=True, startangle=90)
plt.axis('equal') # 원의 형태를 유지 
plt.show()
````

### 연습 문제

````python
bins=[0,20,30,50,70,100]
labels=['미성년자','청년','장년','중년','노년']
explode=(0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]

titanic.dropna(subset=['age'],inplace=True)
titanic['age_group']=pd.cut(titanic.age, bins=bins, labels=labels)
print(titanic['age_group'].value_counts())

plt.pie(titanic['age_group'].value_counts(),labels=titanic['age_group'].value_counts().index,
        explode=explode, colors=colors, shadow= True,
        startangle=90, autopct='%1.lf%%')
plt.axis('equal')
plt.show()
````

# imshow
- 화상(image)데이터처럼 행과 열을 가진 행렬 형태의 2차원 데이터는 imshow 명령을 써서 2차원 자료의 크기를 색깔로 표시

````python
from sklearn.datasets import load_digits
digits = load_digits()
X=digits.images[0]
X
````

- 데이터 수치를 색으로 바꾸는 함수는 칼라맵(color map)이라고 한다.
- 칼라맵은 cmap 인수로 지정한다.
- 사용할 수 있는 칼라맵은 plt.cm의 속성으로 포함되어 있다. 

````python
plt.title('mnist digits; 0')
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.subplots_adjust(left=0.35, right= 0.65, bottom = 0.35, top=0.65)
plt.show()
````

````python
dir(plt.cm)[:10]
````

````python
ig, axes = plt.subplots(1, 4, figsize=(12, 3),
                         subplot_kw={'xticks': [], 'yticks': []})
axes[0].set_title("plt.cm.Blues")
axes[0].imshow(X, interpolation='nearest', cmap=plt.cm.Blues)
axes[1].set_title("plt.cm.Blues_r")
axes[1].imshow(X, interpolation='nearest', cmap=plt.cm.Blues_r)
axes[2].set_title("plt.BrBG")
axes[2].imshow(X, interpolation='nearest', cmap='BrBG')
axes[3].set_title("plt.BrBG_r")
axes[3].imshow(X, interpolation='nearest', cmap='BrBG_r')
plt.show()
````

````python
methods = [
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
    'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
    'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'
]
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
````

# 컨투어 플롯

````python
def f(x,y):
    return (1- x/2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n=256
x=np.linspace(-3, 3, n)
y=np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x,y)
ZZ= f(XX, YY)

plt.title('Contour plots')
plt.contourf(XX, YY, ZZ, alpha =.75, cmap = 'jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
````

# 3D 서피스 플롯
- plot_wireframe()
- plot_surface() 

````python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(XX, YY,ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
````
