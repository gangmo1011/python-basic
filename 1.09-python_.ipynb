# While문

```
while문은 조건식으로만 동작

1. 조건식 판별
2. 조건식이 True인 동안 내내 while문의 바디를 실행
3. 조건식이 False가 되면 while문을 빠져나감

i=0
while i < 100:
    print("Hello World!")
    i += 1
```

### 연습문제

````python
count=int(input('반복할 횟수를 입력하세요: '))
while count >0:
    print('Hello world',count)
    count -= 1
````

````python
i=1
while i < 101:
    if i%2==0:
        print(i,end='\t')
    i += 1
````

````python
num=int(input('구구단을 알려드립니다. 숫자를 입력하세요. '))
i=1
while i <10:
    print(f'{num} * {i} = {num*i}')
    i += 1
````

````python
st_num=int(input('학생의 수 :'))
sum=0
i=1
while i<=st_num:
    score=int(input(f'{i}번 학생의 점수:'))
    sum += score
    i += 1
print(f'평균 : {sum/st_num}점')
````

````python
class_a=(70,60,55,75,95,90,80,80,85,100)
i=0
total=0

while i<len(class_a):
    total +=class_a[i]
    i+=1
print(total/len(class_a))
````

# break문

```
for, while문에서 break는 해당 반복문을 완전히 빠져나감
```

### 연습문제

````python
i =0
total =0
while True:
    total += i
    i +=1
    if i>=101:
        break;
print(total)
````

````python
i=0
while True:
    if i%10==3:
        print(i,end=' ')
    if i>100:
        break
    i+=1

print('\n')

#continue문 사용
i=0
while True:
    if i>100:
        break
    elif i%10!=3:
        i += 1
        continue
    else:
        print(i,end=' ')
        i += 1
````

# continue문

```
continue 는 제어흐름을 유지하되, 밑에 body 부분은 실행하지 않고 반복문 처음으로 돌아감
```

### 연습문제

````python
#for문
for i in range(1,11):
    print(i,end=' ')
    
print('\n')

#while문
i=0
while i<10:
    print(i+1,end=' ')
    i += 1
````

````python
#for문
n=int(input('시작하는 정수 입력 :'))
m=int(input('끝나는 정수 입력 :'))
total=0
for i in range(n,m+1):
    total+=i
print(f'{n}부터 {m}까지의 합은 {total}입니다.')

print('\n')

#while문
n=int(input('시작하는 정수 입력 :'))
m=int(input('끝나는 정수 입력 :'))
total=0

while n<=m:
    total+=n
    n+=1
print(f'{n-m}부터 {m}까지의 합은 {total}입니다.')
````

````python
n=int(input('시작하는 정수 입력: '))
m=int(input('끝나는 정수 입력: '))
total=0
#for문
for i in range(n,m+1):
    if i%2==0:
        total += i*i
print(f'{n}부터 {m}사이의 짝수 제곱의 합은 {total}입니다.')

print('\n')
#while문
n=int(input('시작하는 정수 입력: '))
m=int(input('끝나는 정수 입력: '))
total=0
while n<=m:
    if n%2==0:
        total += n*n
    n += 1
print(f'{n-m}부터 {m}사이의 짝수 제곱의 합은 {total}입니다.')
````

````python
kor=[70,60,55,75,95,90,80,80,85,100]
#for문
total=0
for i in kor:
    total+=i
print(total/len(kor))

#while문
total=0
i=0
while i<len(kor):
    total += kor[i]
    i+=1
print(total/len(kor))    
````
