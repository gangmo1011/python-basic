## 틱 설정하기 
- 축상의 위치 표시 지점을 틱(tick)이라고 함 
- xticks()나 yticks()를 사용

```python
import numpy as np
X= np.linspace(-np.pi, np.pi, 256)
C=np.cos(X)
plt.title('x축과 y축의 tick label 설정')
plt.plot(X,C,c='r',ls='--')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,+1])
plt.show()
```

- 틱 라벨 문자열에는 $$ 사이에 수학 문자식을 넣을 수도 있음
- 다양한 mathematical expressions은 아래 링크를 통해 확인 가능
- https://matplotlib.org/3.1.1/tutorials/text/mathtext.html

plt.show()

## 여러 개의 선 그리기 
- 여러 개의 선을 그리고 싶은 경우에는 x 데이터, y 데이터, 스타일 문자열을 반복하여 인수로 넘김

```python
t= np.arange(0., 5., 0.2)
plt.title('라인 플롯에서 여러개의 선 그리기')
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'gh-')
plt.show()
```

- 하나의 plot()이 아니라 복수의 plot()을 하나의 그림에 겹쳐서 그릴 수도 있다.

```python
plt.title('복수의 plot 명령을 한 그림에서 표현')
style1 = {'c':'b', 'lw':5, 'ls':'--', 'marker':'o','ms':15, 'mec': 'g', 'mew':5, 'mfc':'r'}
style2 = {'c':'k', 'lw':3, 'ls':':', 'marker':'s','ms':10, 'mec': 'm', 'mew':5, 'mfc':'c'}

plt.plot([1,4,9,16],**style1)
plt.plot([9,16,4,1],**style2)

plt.show()
```

## 범례(Legend)
- 라인 플롯을 그리는 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지를 보여주기 위해 legend 명령으로 범례(legend)를 추가 가능
- 범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면 loc 키워드 인수를 사용

```python
X= np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title('legend를 표시한 플롯')
plt.plot(X, C, ls= '--', label='cosine')
plt.plot(X, S, ls=':', label='sine')
plt.legend(loc=2)
plt.show()
```

## Axes
- Figure 객체 안에 여러 개의 플롯을 그리드 형태로 보여야 하는 경우
- Figure 안에 있는 각각의 플롯은 Axes라고 불리는 객체에 속함
- Figure 안에 Axes를 생성하려면 원래 subplot()을 사용하여 명시적으로 Axes 객체를 만들어야 함
- plot()을 바로 사용해도 자동으로 Figure, Axes 객체를 생성함 

- subplot 명령은 세개의 인수를 가지는데, 첫, 두번째 인수는 shape / 세번 째 인수는 몇 번째 오는지를 의미 
- tight_layout() 명령을 실행하면 플롯 간의 간격을 자동으로 맞춰줌

```python
# 처음이 0이 아니라 1
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi * x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped coliation')

ax2=plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('axes 1')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('axes 2')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('axes 3')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('axes 4')

plt.tight_layout()
plt.show()
```

- subplots()에 인수를 두개만 전달하면 복수의 Axes 객체를 동시에 생성 가능
- 이때 2차원 ndarray 형태로 Axes 객체가 반환

```python
fig, axes= plt.subplots(2,2)

np.random.seed(0)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('axes 1')
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('axes 2')
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('axes 3')
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('axes 4')

plt.tight_layout()
plt.show()
```

- twinx() 명령으로 복수의 y축을 가진 플롯을 만들 수도 있음

```python
fig, ax0 = plt.subplots()
ax1= ax0.twinx()
ax0.set_title('2개의 y축 한 figure에서 사용하기')
ax0.plot([10,5,2,9,7], 'r-', label='y0')
ax0.set_ylabel('y0')
ax0.grid(False)

ax1.plot([100,200,220,180,120],'g:',label='y1')
ax1.set_ylabel('y1')
ax1.grid(False)

ax0.set_xlabel('공유되는 x축')
plt.show()
```
